# PANDAS

```python
import pandas as pd
import numpy as np
```

```python
# Create DataFrme

student= {
    'maths':[50,45,48,49,35],
    'english':[42,45,47,44,34],
    'maths':[41,42,43,46,47],
    'computer':[49,45,np.nan,40,np.nan]
    
}
df=pd.DataFrame(student)
df
```

# FILTERING OUT MISSING DATA

```python
# Check Null Values

df.isnull().sum()
```

```python
# drop any row containing missing value

df.dropna()
```

```python
# drop any column containing missing values

df.dropna(axis = 1) 
```

```python
# drop row that are all missing

df.dropna(how = 'any')
```

```python
# fill all missing data with

df1 = df.fillna(0)
df1
```

```python
# modify in-place

df1.fillna('inplace = True')
df1
```

```python
# Only forward fill the 2 missing values in front :

df1.fillna(method = 'ffill', limit = 2)
```
