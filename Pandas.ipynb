## Concat & Merging

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                  index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, 
                  index=[4,5,6,7])
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, )


df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'], 
                    'F': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C0', 'C3', 'C6', 'C7'], 
                    'G': ['D4', 'D5', 'D6', 'D7']}, )
```

```python
df1
```

```python
df2
```

```python
merge_df = pd.merge(df1, df2, on='C', how='inner')
```

```python
merge_df
```

```python
merge_df = pd.merge(df1, df2, on='C', how='left')
```

```python
merge_df
```

```python
merge_df = pd.merge(df1, df2, on='C', how='right')
```

```python
merge_df
```

## loc vs iloc

```python
s = pd.Series(list("abcdef"), index=[49, 48, 47, 0, 1, 2]) 
s
```

```python
s.loc[0]    # value at index label 0
```

```python
s.iloc[3]
```

```python
s.iloc[0]   # value at index location 0
```

```python
s.loc[0:1]  # rows at index labels between 0 and 1 (inclusive)
```

```python
s.iloc[0:1] # rows at index location between 0 and 1 (exclusive)
```

```python
s.iloc[0:2]
```

### What if you sort the Series

```python
s = s.sort_index()
```

```python
s
```

```python
s.iloc[0]
```

```python
s.loc[0]    # value at index label 0
```

```python
s.iloc[0]   # value at index location 0
```

```python
s.loc[0:1]  # rows at index labels between 0 and 1 (inclusive)
```

```python
s.iloc[0:1] # rows at index location between 0 and 1 (exclusive)
```
